```
lr = db_airbnb.listingsAndReviews
```

```
res = list(lr.find({
              'address.market': 'Montreal',
              'bedrooms': {"$gt": 5},
              'amenities': {"$all": ['Laptop friendly workspace', 'Dishwasher']}
        }))
```

```
len(res)
```

```
# 40.748106, -74.012900
# 40.702011, -73.973594
```

```
res = list(lr.find({
              'address.market': 'New York',
              'bedrooms': {"$gt": 3},
              'address.location.coordinates.1': {'$gt': 40.702011, '$lt': 40.748106},
              'address.location.coordinates.0': {'$gt': -74.012900, '$lt': -73.973594},
        }))
```

```
len(res)
```

```
res[0]['address']['location']['coordinates']
```

```
latlon_list = [p['address']['location']['coordinates'] for p in res]
latlon_list
```

```
import folium
```

```
folium_map = folium.Map(location=[40.719283, -73.999576], zoom_start=20)
for coords in latlon_list:
  folium.Marker([coords[1], coords[0]]).add_to(folium_map)
folium_map
```
