## U = Update


```
db.test.update_one({'value': 7},
                   {'$inc': {'value': 100}})

list(db.test.find())
```

```
db.test.update_many({'even': True},
                   {'$inc': {'value': 100}})

list(db.test.find())
```

## D = Delete

```
db.test.delete_many({'even' : False})
list(db.test.find())
```
