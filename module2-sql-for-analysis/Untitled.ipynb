```python
# open a connection and cursor to sqlite db
sl_conn = sqlite3.connect('rpg_db (1).sqlite3')
sl_curs = sl_conn.cursor()
```

```python
# Get all the character data from the table
characters = sl_curs.execute('SELECT * FROM charactercreator_character;').fetchall()

characters
```

```python
# Get table metadata
sl_curs.execute('PRAGMA table_info(charactercreator_character);').fetchall()
```

```python
# create matching schema for postgres db
create_character_table = """
    CREATE TABLE charactercreator_character (
      character_id SERIAL PRIMARY KEY,
      name VARCHAR(30),
      level INT,
      exp INT,
      hp INT,
      strength INT,
      intelligence INT,
      dexterity INT,
      wisdom INT
    );
"""
```

```python
pg_curs.execute(create_character_table)
```

```python
# Now we can look at what tables are here
show_tables = """
SELECT 
 *
FROM
   pg_catalog.pg_tables
WHERE
    schemaname != 'pg_catalog'
AND schemaname != 'information_schema';
"""
pg_curs.execute(show_tables)
pg_curs.fetchall()
```

```python
# Test grabbing a single row of values
str(characters[0][1:])
```

```python
# code for inserting row into table
insert = """
INSERT INTO charactercreator_character
(name, level, exp, hp, strength, intelligence, dexterity, wisdom)
VALUES """
```

```python
# insert it all into new table
for character in characters:
    insert_character = insert + str(character[1:]) + ';'
    pg_curs.execute(insert_character)
```

```python
# Look inside and find all the data
pg_curs.execute('SELECT * FROM charactercreator_character')
pg_curs.fetchall()
```

```python
# get a list of the data from postgres
pg_curs.execute('SELECT * FROM charactercreator_character;')
pg_characters = pg_curs.fetchall()
```

```python
# Close connection and commit to secure changes
pg_curs.close()
pg_conn.commit()
```

```python
# Test that data is the same 
for character, pg_character in zip(characters, pg_characters):
    assert character == pg_character
```
