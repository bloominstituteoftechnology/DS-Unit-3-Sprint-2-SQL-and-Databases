```python
create_character_table = '''
  CREATE TABLE character_table (
  character_id SERIAL PRIMARY KEY,
  name VARCHAR(30),
  level INT,
  exp INT,
  hp INT,
  strength INT,
  intelligence INT,
  dexterity INT,
  wisdom INT
);
'''
pg_cursor.execute(create_character_table)
```
